```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the data into python

```python
data=pd.read_csv('Week_7_Activity_data.csv')
data
```

```python
data.info()
```

# 2. Check for null values

```python
data.isnull().sum()
```

**Findings:**
-There is no null value present in the dataset. 

# 3. Examine the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ and comment on your observations.

```python
data1 = data[['Hardness', 'Sulfate', 'Organic_carbon']]
data1.describe()
```

```python
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.histplot(data1['Hardness'],bins=10,kde=True)
plt.subplot(2,2,2)
sns.histplot(data1['Sulfate'],bins=10,kde=True)
plt.subplot(2,2,3)
sns.histplot(data1['Organic_carbon'],bins=10,kde=True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=data1)
```

```python
corrmatrix = data1.corr()
plt.subplots(figsize=(10,8))
sns.heatmap(corrmatrix,vmin=-0.3,vmax=0.3,annot=True,linewidths=0.2,cmap='Blues')
```

**Findings**

- The parameters Hardness, Sulfate and Organic_carbon are almost normally distributed.
- There are few outliers present for all thre parameters.
- The three parameters are weakly correlated to each other
