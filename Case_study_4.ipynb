```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
data=pd.read_csv('Sales_add.csv')
data
```

```python
a1=data['Sales_before_digital_add(in $)'].mean()
a2=data['Sales_After_digital_add(in $)'].mean()
print('Average sales before digital add is %0.3f' %a1)
print('Average sales after digital add is %0.3f' %a2)
```

```python
ttest, p=stats.ttest_rel(data['Sales_After_digital_add(in $)'],data['Sales_before_digital_add(in $)'],alternative='greater')
print('Null Hypothesis: Average sales before digital add and after digital add is same')
print('Alternative Hypothesis: Average sales before digital add is less than the average sales after digital add')
print('p value: %0.12f' %p)
if p<=0.05:
    print('Since, p value, %0.12f is less than 0.05, we reject Null Hypothesis. Hence, we can say that average sales before digital add is less than the average sales after digital add differs' %p)
else:
    print('Since, p value, %.12f is more than 0.05, we accept Null Hypothesis. Hence, we can say that the average sales before digital is greater than the average sales after digital add'%p)
```

```python
datas=data[['Region','Manager']]
datas
```

```python
freq=datas[['Region','Manager']].groupby('Region').count()
```

```python
stat,p=stats.chisquare(freq)
print('Null Hypothesis: There is no relationship between Regions and Managers')
print('Alternative Hypothesis: There is a relationship between Regions and Managers')
print('p value: %0.3f' %p)
if p<=0.05:
    print('Since, p value, %0.3f is less than 0.05, we reject Null Hypothesis. Hence, we can say that there is a relationship between Regions and Managers' %p)
else:
    print('Since, p value, %.3f is more than 0.05, we accept Null Hypothesis. Hence, we can say that there is no relationship between Regions and Managers'%p)
```
